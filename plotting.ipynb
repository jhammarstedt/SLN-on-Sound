```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sym10snl = pd.read_json('results/cifar10_symmetric/training_log_base.json',)
sym10 = pd.read_json('results/cifar10_symmetric_noSLN/training_log.json',)
asym10snl = pd.read_json('results/cifar10_asymmetric/training_log.json',)
asym10 = pd.read_json('results/cifar10_asymmetric_noSLN/training_log.json',)
sym100snl = pd.read_json('results/cifar100_symmetric/training_log.json',)
asym100snl = pd.read_json('results/cifar100_asymmetric/training_log.json',)
sym100 = pd.read_json('results/cifar100_symmetric_noSLN/training_log.json',)
asym100 = pd.read_json('results/cifar100_asymmetric_noSLN/training_log.json',)
fsd_sym_snl = pd.read_json('results/fsd_symmetric/training_log.json',)
fsd_sym = pd.read_json('results/fsd_symmetric_noSLN_14/training_log.json',)
fsd_sym = fsd_sym.append( pd.read_json('results/fsd_symmetric_noSLN_15/training_log.json',), ignore_index=True)
```

```python
def plot(data, labels, title='CIFAR-10 test accuracy', ylabel='Accuracy'):
    for i in range(len(data)):
        plt.plot(data[i], label=labels[i])
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.xlabel('Epoch')
    plt.ylabel(ylabel)
    plt.xticks(np.arange(0, data[0].shape[0], 3.0))
    plt.savefig(f'./plots/{title}.png')
    plt.show()
```

```python
# plt.plot(sym10snl.test_acc)
# plt.plot(sym10.test_acc)
plot([sym10snl.test_acc, sym10.test_acc, asym10snl.test_acc, asym10.test_acc], 
     ['Symmetric with SLN', 'Symetric without SLN', 'Asymmetric with SLN', 'Asymetric without SLN'], 
     title='CIFAR-10 test accuracy'
    )
```

```python
plot([sym10snl.train_acc, sym10.train_acc, asym10snl.train_acc, asym10.train_acc], 
     ['Symmetric with SLN', 'Symetric without SLN', 'Asymmetric with SLN', 'Asymetric without SLN'], 
     title='CIFAR-10 train accuracy'
    )
```

```python
plot([sym10snl.train_loss, sym10.train_loss, asym10snl.train_loss, asym10.train_loss], 
     ['Symmetric with SLN', 'Symetric without SLN', 'Asymmetric with SLN', 'Asymetric without SLN'],
     title='CIFAR-10 train loss',
     ylabel='CE loss'
    )
```
